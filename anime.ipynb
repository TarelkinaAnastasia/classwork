```python
import numpy as np
import pandas as pd
df=pd.read_csv("anime.csv")
```

```python
df.columns
```

```python
df['Episodes'] = df['Episodes'].replace('?', -1).replace(-1, np.nan)
df['Voters'] = df['Voters'].str.replace(",", "")
df = df.astype({'Episodes': 'float32', 'Title': 'string', 'Production': 'string', 'Source': 'string', 'Genre': 'string', 'Airdate': 'string','Voters': 'int32', 'Rating': 'float32', 'Theme': 'string'})
print(df.dtypes)
```

```python
columns_ = ['title', 'production', 'episodes', 'source', 'genre', 'airdate',
       'rating', 'voters', 'theme']
  
# Set the index
df.columns = columns_
  
# Print the DataFrame
df
```

```python
df.dropna()
```
